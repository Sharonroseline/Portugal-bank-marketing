### 2)b

```python
#Encoding Categorical Variables using Label Encoding
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
label_df=['job','marital','education','default','housing','loan','month','day_of_week','poutcome','y']
df[label_df]=df[label_df].apply(label.fit_transform)
```

```python
df = pd.get_dummies(df, columns=['contact'])
df.head()
```

### 2)c

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
```

```python
X = df.drop(['y'],axis=1)
Y = df[['y']]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.30, random_state=0)
```

```python
rf=SelectFromModel(RandomForestClassifier(n_estimators = 100))
rf.fit(X_train,Y_train)
```

```python
rf.get_support()
```

```python
selected_feat= X_train.columns[(rf.get_support())]
len(selected_feat)
```

```python
print(selected_feat)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

### 3)d

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
neighbors=list(range(1,40,2))
cv_scores=[]
for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn,X,Y,cv=10,scoring='accuracy')
    cv_scores.append(scores.mean())
```

```python
MSE=[1-x for x in cv_scores]
MSE
```

```python
optimal_k=neighbors[MSE.index(min(MSE))]
optimal_k
```

```python
NNH = KNeighborsClassifier(n_neighbors=39  ,metric='euclidean' )
```

```python
NNH.fit(X_train, Y_train)
```

```python
predicted_labels = NNH.predict(X_test)
NNH.score(X_train, Y_train)
```

```python
predicted_labels = NNH.predict(X_test)
NNH.score(X_test, Y_test)
```
